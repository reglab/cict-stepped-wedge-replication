Reproducing the main figures (difference-in-differences time series) from the paper, *Automated vs. Manual Case Investigation and Contact Tracing for Pandemic Surveillance: Evidence from a Stepped Wedge Cluster Randomized Trial*.

Note: while this notebook replicates our main results with aggregate data, the original paper performed analyses at the individual level. 


```r
library(tidyverse)
library(patchwork)
library(scales)
```

```r
panel_data <- read_csv('./panel_data.csv')
```

```r
# Helper function that takes the zip x week panel and converts 
# it into the cluster x week panel with seperate overflow values
format_data <- function(grouped_tb, outcome) {
    
    auto <- paste0(outcome, '_assigned_automated')
    manual <- paste0(outcome, '_assigned_manual')
    
    # Create crossover-week panel for the compliers
    compliers <- grouped_tb %>% 
        # Use complier completion rates
        mutate(
            completion_rate = ifelse(condition == 'Automated CICT',
                                                  !!sym(auto),
                                                  !!sym(manual)),
            # Same with case numbers
            n = ifelse(condition == 'Automated CICT',
                       n_assigned_automated,
                       n_assigned_manual)) %>% 
        summarize(
            completion_rate = weighted.mean(completion_rate,n),
            n = sum(n),
            .groups = 'drop')
    # Create a seperate time series for the overflow cases
    overflow <- grouped_tb %>% 
        mutate(crossover = -1) %>% 
        # Overflow = ZIP in manual CICT but assigned automated CICT
        filter(condition == 'Manual CICT') %>% 
        summarize(
            n = sum(n_assigned_automated),
            completion_rate = weighted.mean(!!sym(auto), n_assigned_automated),
            .groups = 'drop')
    
    # Bind rows and return 
    bind_rows(compliers,overflow)
}
```

```r
# Helper lists
colors <- c('Grey', '#6FB1F7', '#4771CC', '#1F30A1', 'Red')

labels <- c('Overflow', 'Early', 'Medium', 'Late', 'Never')

summary_cols <- c('perc_filled',
      'has_race_ethnicity',
      'has_employer',
      'has_gender',
      'has_sexual_orientation',
      'has_language',
      'has_signs_or_symptons',
      'has_travel',
      'has_attended_gathering',
      'has_able_to_self_isolate',
      'has_num_contacts_generated',
      'has_congregate_setting')

outcome_vars <- paste0('mean_', summary_cols)

plain_outcome_vars <- c(
    '% Fields Known',
    'Race/Ethnicity Known',
    'Employer Known',
    'Gender Known',
    'Sexual Orientation Known',
    'Language Known',
    'Signs/Symptoms Known',
    'Travel History Known',
    'Gathering History Known',
    'Ability to Isolate Known',
    'Contacts Generated Known',
    'Congregate Setting Known')
```

## Figure 2

Time series for a sample of outcomes

```r
# overall completion rate (ocr) - Compliers
g.2 <- panel_data %>% 
    group_by(week,crossover) %>% 
    format_data('mean_perc_filled') %>%
    ggplot(aes(x = week, y = completion_rate, group = crossover, color = as.factor(crossover))) + 
    geom_line(alpha = 0.75) + 
    geom_point(alpha = 0.75, aes(size = n)) +
    scale_y_continuous(labels = percent, limits = c(0,1)) +
    labs(x = '', y = '% Fields Known', shape = 'Stepdown', color = 'Stepdown Date', size = 'Case Count') + 
    scale_color_manual(values = colors, labels = labels) +
    theme_minimal() 

# Go through the main columns, format data, create new chart and append to existing chart
for (i in c(7,9,3)) {
    col <- outcome_vars[i]
    name <- plain_outcome_vars[i]
    
    new.g.2 <- panel_data %>% 
        group_by(week,crossover) %>% 
        format_data(col) %>%
        ggplot(aes(x = week, y = completion_rate, group = crossover, color = as.factor(crossover))) + 
        geom_line(alpha = 0.75) + 
        geom_point(alpha = 0.75, aes(size = n)) +
        scale_y_continuous(labels = percent, limits = c(0,1)) +
        labs(x = '', y = name, shape = 'Stepdown', color = 'Stepdown Date', size = 'Case Count') + 
        scale_color_manual(values = colors, labels = labels) +
        theme_minimal() 
    
    g.2 <- g.2 + new.g.2
}

# Render and format
g.2 + 
    plot_layout(guides = 'collect') & 
    theme(legend.position = 'right',
          panel.grid.minor.x = element_blank(),
          panel.grid.major.x = element_blank())
```
